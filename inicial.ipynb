```python
import pandas as pd

# Carregar o conjunto de dados
tabela = pd.read_csv("dataSet.csv", sep=';')

# Selecionar apenas as colunas desejadas, as outras não eram relevantes para o propósito da análise
colunas_desejadas = ['ANO', 'NOMEESCOLA', 'ANOENSINO', 'SERIE', 'TURNO', 'SEXOALUNO', 'DATAALUNONASC', 'IDADE', 'SITUACAO', 'SITUDESC']
tabela_selecionada = tabela[colunas_desejadas]

# Exibir o DataFrame com as colunas desejadas
display(tabela_selecionada)
```

```python
print("\nInformações do dataframe selecionado:")
tabela_selecionada = tabela_selecionada.dropna()
display(tabela_selecionada.info())
```

```python
# Verificar valores ausentes na tabela inteira
print("\nValores ausentes por variável:")
print(tabela.isnull().sum())
```

```python
# Análise das estatísticas descritivas da coluna idade
idade_coluna = tabela['IDADE']
estatisticas_idade = idade_coluna.describe().round(2)
print(estatisticas_idade)
```

```python
# Verificar a quantidade de registros únicos em cada variável categórica
print("\nQuantidade de registros únicos em variáveis categóricas:")
print(tabela_selecionada.nunique())
```

```python
print("\nSituação do alunos:")
print(tabela['SITUDESC'].value_counts().round(2))
```

```python
# Calcular estatísticas descritivas da coluna de idade para todos os alunos
estatisticas_idade_geral = tabela['IDADE'].describe().round(2)

# Exibir as estatísticas descritivas da idade para todos os alunos
print("\nEstatísticas descritivas da idade para todos os alunos:")
print(estatisticas_idade_geral)

# Filtrar o DataFrame para incluir apenas os alunos que desistiram
alunos_desistentes = tabela[tabela['SITUDESC'] == 'DESIST/DEIXOU FREQUE']

# Calcular estatísticas descritivas da coluna de idade dos alunos desistentes
estatisticas_idade_desistentes = alunos_desistentes['IDADE'].describe().round(2)

# Exibir as estatísticas descritivas da idade dos alunos desistentes
print("\nEstatísticas descritivas da idade dos alunos desistentes:")
print(estatisticas_idade_desistentes)
```

```python
# Identificação de outliers usando o método IQR (Intervalo Interquartil)
Q1 = tabela['IDADE'].quantile(0.25)
Q3 = tabela['IDADE'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Remoção de outliers
tabela_sem_outliers = tabela[(tabela['IDADE'] >= limite_inferior) & (tabela['IDADE'] <= limite_superior)]

# Exibir estatísticas descritivas da idade após remoção de outliers
print(tabela_sem_outliers['IDADE'].describe().round(2))
```

```python
import matplotlib.pyplot as plt

print("Perfil demográfico dos alunos")

# Distribuição de Idade
plt.figure(figsize=(10, 6))
plt.hist(tabela['IDADE'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição de Idade dos Alunos')
plt.xlabel('Idade')
plt.ylabel('Número de Alunos')
plt.grid(True)
plt.show()

# Distribuição de Sexo
plt.figure(figsize=(8, 5))
tabela['SEXOALUNO'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribuição de Alunos por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Número de Alunos')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

# Modificar os rótulos da coluna 'SERIE'
tabela['SERIE'] = tabela['SERIE'].replace('EDUCACAO JOVENS E ADULTOS', 'EJA')

# Plotar o gráfico com os rótulos modificados
plt.figure(figsize=(8, 5))
tabela['SERIE'].value_counts().sort_index().plot(kind='bar', color='lightblue')
plt.title('Distribuição de Alunos por Série')
plt.xlabel('Série')
plt.ylabel('Número de Alunos')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

```

```python
# Filtrar o DataFrame para incluir apenas os alunos que desistiram
alunos_desistentes = tabela[tabela['SITUDESC'] == 'DESIST/DEIXOU FREQUE']

# Contar o número de desistências por série
desistencias_por_serie = alunos_desistentes['SERIE'].value_counts()

print("\nDesistências por Série:")
print(desistencias_por_serie)
```

```python
import matplotlib.pyplot as plt

# Filtrar o DataFrame para incluir apenas os alunos que desistiram e têm até 20 anos de idade
alunos_desistentes_ate_20 = tabela[(tabela['SITUDESC'] == 'DESIST/DEIXOU FREQUE') & (tabela['IDADE'] <= 20)]

# Calcular o número de desistências por idade e sexo
desistencias_por_idade_sexo = alunos_desistentes_ate_20.groupby(['IDADE', 'SEXOALUNO']).size().unstack()

# Plotar o gráfico de barras para desistências por idade e sexo
desistencias_por_idade_sexo.plot(kind='bar', figsize=(10, 6))
plt.title('Desistências por Idade e Sexo (Até 20 anos)')
plt.xlabel('Idade')
plt.ylabel('Número de Desistências')
plt.xticks(rotation=45)
plt.legend(title='Sexo', loc='upper right')
plt.grid(axis='y')
plt.show()

```

```python
import pandas as pd
import folium

# Carregar o conjunto de dados
tabela = pd.read_csv("dataSet.csv", sep=';')

# Selecionar as colunas
colunas_desejadas = ['NOMEBAIRRO', 'SITUDESC']
tabela_selecionada = tabela[colunas_desejadas]

# Filtrar o DataFrame para incluir apenas os alunos que desistiram nos bairros de interesse
alunos_desistentes = tabela_selecionada[tabela_selecionada['SITUDESC'] == 'DESIST/DEIXOU FREQUE']

# Contagem de desistências por bairro
desistencias_por_bairro = alunos_desistentes['NOMEBAIRRO'].value_counts()

# Coordenadas para os bairros de interesse
coordenadas_bairros = {
    'ESTANCIA': (-8.085244409718053, -34.93012041204825),
    'IPUTINGA': (-8.036625544717898, -34.93554927751233),
    'JARDIM SAO PAULO': (-8.081547856224457, -34.94432580422609),
    'COHAB': (-8.119167604544865, -34.953142399160335),
    'IBURA': (-8.11095277783638, -34.93557486825973)
}

# Criar um mapa centrado em Recife
mapa = folium.Map(location=[-8.0476, -34.8770], zoom_start=12)

# Adicionar marcadores para cada bairro onde houve desistências
for bairro, total_desistencias in desistencias_por_bairro.items():
    if bairro in coordenadas_bairros:
        coord_bairro = coordenadas_bairros[bairro]
        folium.Marker(location=coord_bairro, popup=f"Bairro: {bairro}<br>Desistências: {total_desistencias}",
                      icon=folium.Icon(color='red' if bairro == 'ESTANCIA' else 'blue')).add_to(mapa)

# Adicionar título ao mapa
titulo_mapa = f"Mapa de Calor Aproximado de Desistências em Recife"
folium.TileLayer('cartodbpositron').add_to(mapa)  # Adicionar camada de fundo

# Salvar o mapa como um arquivo HTML
mapa.save('mapa_calor_desistencias_recife_aprox.html')

# Exibir o mapa
mapa

```

```python
import pandas as pd
import folium
import dash
from dash import html, dcc
import plotly.express as px

# Carregar o conjunto de dados
tabela = pd.read_csv("dataSet.csv", sep=';')

# Filtrar o DataFrame para incluir apenas os alunos que desistiram e têm até 20 anos de idade
alunos_desistentes_ate_20 = tabela[(tabela['SITUDESC'] == 'DESIST/DEIXOU FREQUE') & (tabela['IDADE'] <= 20)]

# Contagem de desistências por bairro
desistencias_por_bairro = tabela[tabela['SITUDESC'] == 'DESIST/DEIXOU FREQUE']['NOMEBAIRRO'].value_counts()

# Coordenadas para os bairros de interesse
coordenadas_bairros = {
    'ESTANCIA': (-8.085244409718053, -34.93012041204825),
    'IPUTINGA': (-8.0476, -34.9500),
    'JARDIM SAO PAULO': (-8.0000, -34.9000),
    'COHAB': (-8.0700, -34.9300),
    'IBURA': (-8.1000, -34.9500)
}

# Criar um mapa centrado em Recife
mapa = folium.Map(location=[-8.0476, -34.8770], zoom_start=12)

# Adicionar marcadores para cada bairro onde houve desistências
for bairro, total_desistencias in desistencias_por_bairro.items():
    if bairro in coordenadas_bairros:
        coord_bairro = coordenadas_bairros[bairro]
        folium.Marker(location=coord_bairro, popup=f"Bairro: {bairro}<br>Desistências: {total_desistencias}",
                      icon=folium.Icon(color='red' if bairro == 'ESTANCIA' else 'blue')).add_to(mapa)

# Configuração do Dash
app = dash.Dash(__name__)

# Layout do Dashboard
app.layout = html.Div([
    html.H1("Mapa de Desistências por Bairro"),

    # Mapa de Desistências por Bairro
    html.Div([
        html.Iframe(id='mapa', srcDoc=mapa._repr_html_(), width='100%', height='600')
    ])
])

if __name__ == '__main__':
    app.run_server(debug=True, port=5000)
```
